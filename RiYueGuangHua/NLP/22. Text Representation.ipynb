### 分词

```python
s = 'Life is not easy for any of us.We must work,and above all we must believe in ourselves.We must believe that each one of us is able to do some thing well.And that we must work until we succeed.'
```

```python
s
```

```python
import string
```

```python
string.punctuation
```

```python
for c in string.punctuation:
    s = s.replace(c, ' ').lower()
```

```python
s
```

分词方式（一）

```python
list(s)
```

分词方式（二）

```python
s.split()
```

分词方式（三）: n-gram

向量化： one-hot   emdeding

```python
np.unique(s.split())
```

```python
vocab = dict((word, index) for index, word in enumerate(np.unique(s.split())))
```

```python
vocab
```

```python
s = [vocab.get(w) for w in s.split()]
```

```python
s
```

```python
len(s), len(vocab)
```

```python
b = np.zeros((len(s), len(vocab)))
```

```python
for index, i in enumerate(s):
    b[index, i] = 1
```

```python
b[:5]
```

词嵌入表示

```python
em = torch.nn.Embedding(len(vocab), 20)
```

```python
len(s)
```

```python
s_em = em(torch.LongTensor(s))
```

```python
s_em.shape
```

```python
s_em
```

嵌入层还用于所有类型的分类输入，而不只是为自然语言进行词嵌入。

例如，如果你要为超市里多个商品做推荐，则最好对商品名称做词嵌入，而不是将它们作为单热编码向量输送给网络。
